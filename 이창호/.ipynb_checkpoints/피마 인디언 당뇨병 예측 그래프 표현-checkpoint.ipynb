```python
import pandas as pd
```

```python
# pandas read
df = pd.read_csv('pima-indians-diabetes.csv', 
                 names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])
print(df)
```

```python
# head 함수로 0~4번까지 확인
print(df.head(5))
```

```python
# info() 함수로 데이터의 전반적인 정보 확인
print(df.info())
```

```python
# describe() 함수로 좀 더 자세한 정보 확인
print(df.describe())
```

```python
# 데이터의 일부분만 추출
# pandas.DataFrame.loc[행 정보, 열 정보]
print(df[['pregnant', 'class']])
```

```python
# groupby() 함수로 pregnant 정보를 기준으로 새 그룹 생성
# as_index=False로 pregnant 정보 옆에 새로운 인덱스 생성
# mean() 함수로 평균 계산
# sort_values() 함수로 pregnant의 정보를 오름차순(ascending)으로 정렬
print(df[['pregnant', 'class']].groupby(['pregnant'], as_index=False).mean().sort_values(by='pregnant', ascending=True))
```

```python
# matplotlib 라이브러리는 시각화를 위해 사용
# seaborn 라이브러리는 정교한 그래프를 그리기 위해 사용
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 그래프 크기 설정
plt.figure(figsize=(12,12))
```

```python
# heatmap() 함수로 각 항목 간의 상관관계를 표시
# heatmap() 함수는 두 항목씩 짝을 지은 뒤 두 항목이 전혀 다른 패턴으로 변화하면 0 비슷한 패턴으로 변화할수록 1에 가까운 값을 출력
sns.heatmap(df.corr(), linewidths=0.1, vmax=0.5, cmap=plt.cm.gist_heat, linecolor='white', annot=True)
```

```python
# class가 0일때와 1일때의 plasma 정보 확인
grid = sns.FacetGrid(df, col='class')
grid.map(plt.hist,'plasma', bins=10)
plt.show()
```
